# Identifier

#### •	A name in python program is called identifier


## Rules to create identifier
#### 1.	Identifier name must be sequence of upper-case(A-Z) lower-case(a-z) digits(o-9) and underscore (_).
#### 2.	Identifier can-not start with digit.
#### 3.	Python is a case sensitive language so upper case and lower case is different.
#### 4.	Keyword can not be used as a keyword.
#### 5.	No length limit for creating python identifier.

# Variable
#### •	A variable helps to store data
#### •	When we assign the data then the variable is created
#### •	We do not need to declare the type while creating the variable.
#### •	In python type concept is available but we do not need to declare explicitly


```python
var=10
print(var)
```

#### •	We can change the type after creating the variable

```python
x="string"
print(x)
x=10
print(x)
```

#### •	We can get the data type of a variable with the type () function

```python
x="string"
print(x)
print(type(x))
x=10
print(x)
print(type(x))
```

####  We can get the address of the any data with the id() function

```python
v=123
print(id(v))
```

#### •	when we modify the value of existing variable then the compiler store modified data in new place and gives the address to the variable so we can access the modified data.

```python
h="machine learnhing"
print(h)
print(id(h))
h=12
print(h)
print(id(h))
```

#### •	if we created multiple variables to access same data then all variable access the same address but the data must be in (-5 from 256)

```python
M1=57
M2=57
M3=57
print(M1)
print(M2)
print(M3)
# if we want to print many variable in single line then we use this sytax
print(M1,M2,M3)
print(M1, id(M1), M2, id(M2), M3, id(M3))
```
